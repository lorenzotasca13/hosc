```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
term_cold=np.loadtxt('../main/termalisation_cold.txt')
term_cold=term_cold.transpose()
term_hot=np.loadtxt('../main/termalisation_hot.txt')
term_hot=term_hot.transpose()
```

```python
plt.plot(term_cold[0], term_cold[1], 'r', label='A freddo')
plt.plot(term_hot[0], term_hot[1], 'b', label='A caldo')
plt.xlabel('Numero di sweep')
plt.ylabel('Azione')
plt.title('Termalizzazione')
plt.legend()
plt.show()
```

L'accettanza media a freddo è: 0.542266
L'accettanza media a caldo è: 0.542109
